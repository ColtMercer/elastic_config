# Elastic Training with Kabana
This training material is to familiarize you with the Elastic Search Queries using Kibana and the Elastic API. The dataset used in this training is the collection of all Stehpen King Novels downloaded from Kaggle. The training will be used to answer 10 trivia questions about Stephen King Novels using advanced search queries in Kibana and via API. Each question will be answered showing how to find answer using Kibana UI, Python Requests, Python ElasticSearch Client and converting the query response to a pandas dataframe.

## Environment:
The source code will build a docker environment with the following containers:
- 3 x ElasticSearch Indexer Nodes
- 1 x Kibana (Elastic UI and Visualization Tool)
- 1 x MongoDB Server
- 1 x MondoDB Express Server
- 1 x Python container that will create the data set and adding to elastic indexers

## Setup
1. Install Docker Desktop (Windows/Mac) or Docker Engine (Linux)
  - https://www.docker.com/products/docker-desktop
2. Clone the repository
```bash
git clone <repo url>
```
3. Run the following command to build the environment
```bash
docker-compose up -d
```
4. Open Kibana in your browser: http://localhost:5601
```note
Default credentials for elastic API
  - Username: elastic
  - Password: changeme
```
5. Open MongoDB Express in your browser: http://localhost:8081 (This trainging will not cover MongoDB)
```note
Credentials Set in the docker-compose.yml file
  - Username: devuser
  - Password: devpasswd
```
6. Open Jupyter Notebook using your preferred method
7. Install the required python packages
Using pip
```bash
pip install -r requirements.txt
```
using conda
```bash
conda install --file requirements.txt
```

## Troubleshooting
If data is not created, check the logs of the data_creator container. It may be that the data is not being created because the elasticsearch container is not ready yet. In that case, restart the data_creator container. There are sleep timers to minimize but slower computers may need to increase the sleep timers.

## Quiz: Question 1
### What is the total number of Stephen King Novels in the dataset?*
To answer this question we won't need to use any advanced methods. Just a simple count of the documents in the index. The documents in Elastic are JSON objects created from the data set. The JSON structure is as followed.
```json
{
    "author": "Stephen King",
    "year": year,
    "title": title,
    "contents": book_contents
}
```
The index name is stephen_king

### Kibana UI
1. Open Kibana in your browser: http://localhost:5601
2. Click on the "Discover" tab on the left side of the screen
3. Click Index Management
4. You will see the total number of documents in the index
